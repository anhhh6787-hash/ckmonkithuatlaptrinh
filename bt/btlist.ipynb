```python
#tính tổng list
n = int(input("Nhập số lượng phần tử: "))
a = []

for i in range(n):
    x = int(input("Nhập số: "))
    a.append(x)

print("Tổng =", sum(a))
```

```python
#tìm max,min
n = int(input("Nhập số lượng phần tử: "))
a = []

for i in range(n):
    x = int(input("Nhập số: "))
    a.append(x)

print("Lớn nhất =", max(a))
print("Nhỏ nhất =", min(a))
```

```python
#đếm số chẵn,số lẻ
n = int(input("Nhập số lượng phần tử: "))
a = []

for i in range(n):
    x = int(input("Nhập số: "))
    a.append(x)

chan = 0
le = 0

for x in a:
    if x % 2 == 0:
        chan += 1
    else:
        le += 1

print("Số chẵn:", chan)
print("Số lẻ:", le)
```

```python
#viết hoa toàn bộ tên
n = int(input("Nhập số lượng tên: "))
names = []

for i in range(n):
    ten = input("Nhập tên: ")
    names.append(ten.upper())

print("Danh sách tên viết hoa:", names)
```

```python
# list >10
n = int(input("Nhập số lượng phần tử: "))
a = []

for i in range(n):
    x = int(input("Nhập số: "))
    a.append(x)

b = []   
for x in a:
    if x > 10:
        b.append(x)

print("Các số > 10:", b)
```
