```python
#Đếm chữ hoa, chữ thường, chữ số
s = input("Nhập chuỗi: ")

upper_count = 0
lower_count = 0
num_count = 0

for ch in s:
    if ch.isupper():
        upper_count += 1
    elif ch.islower():
        lower_count += 1
    elif ch.isdigit():
        num_count += 1

print("Số chữ hoa:", upper_count)
print("Số chữ thường:", lower_count)
print("Số chữ số:", num_count)
```

```python
#Đếm số lần xuất hiện của một ký tự
chuoi = input("Nhập chuỗi: ")
kitu = input("Nhập ký tự cần đếm: ")

count = 0
for i in chuoi:
    if i == kitu:
        count += 1

print("Số lần xuất hiện:", count)
```

```python
#Tách số và chữ
s = input("Nhập chuỗi: ")

letters = ""
numbers = ""

for ch in s:
    if ch.isalpha():
        letters += ch
    elif ch.isdigit():
        numbers += ch

print("Chuỗi chữ:", letters)
print("Chuỗi số:", numbers)
```

```python
s = input("Nhập chuỗi: ")

has_letter = False
has_num = False

for i in s:
    if i.isalpha():
        has_letter = True
    elif i.isdigit():
        has_num = True

if has_letter and has_num:
    print("Hợp lệ")
else:
    print("Không hợp lệ")
```
