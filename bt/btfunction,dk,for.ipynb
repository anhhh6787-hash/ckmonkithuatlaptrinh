```python
# Ví dụ 1:
# Viết đoạn code kiểm tra một số nguyên n
# Nếu n là số chẵn -> in "Even"
# Nếu n là số lẻ -> in "Odd"
n = int(input("Enter a number: "))
if n % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Ví dụ 2:
# Cho trước điểm số (0-100).
# Nếu điểm >= 85 -> "Excellent"
# Nếu điểm >= 70 -> "Good"
# Ngược lại -> "Needs improvement"
score = int(input("Enter score: "))
if score >= 85:
    print("Excellent")
elif score >= 70:
    print("Good")
else:
    print("Needs improvement")
```

```python
n = int(input("Nhập số lượng phần tử: "))

chan = 0
le = 0

for i in range(n):
    x = int(input("Nhập số thứ {}: ".format(i+1)))
    if x % 2 == 0:
        chan += 1
    else:
        le += 1

print("Số chẵn:", chan)
print("Số lẻ:", le)
```

```python
# Chương 4
# Bài 1:
# Viết chương trình tính tiền điện, sử dụng một hàm có tên compute_bill nhận hai tham số:
# kwh: số kWh điện đã sử dụng
# rate: đơn giá (tiền/kWh)
# Nếu số kWh vượt quá 100, phần vượt sẽ được tính thêm 20% phụ phí
#                                        Giải:
def compute_bill(kwh, rate):
    if kwh > 100:
        bill = (100 * rate) + ((kwh - 100) * rate * 1.2)
    else:
        bill = kwh * rate
    return bill
try:
    kwh = float(input("Nhập số kWh đã sử dụng: "))
    rate = float(input("Nhập đơn giá (tiền/kWh): "))
    total_bill = compute_bill(kwh, rate)
    print("Tổng tiền điện: ", round(total_bill, 2))
except:
    print("Vui lòng nhập giá trị số hợp lệ.")
```

```python
# Bài 2:
# Viết hàm computeage nhận vào một số tuổi và trả về phân loại sau:
#  Age                  | Category        |
#  dưới 0 hoặc trên 120 | `"Invalid age"` |
#  0–5                  | `"Toddler"`     |
#  6–12                 | `"Child"`       |
#  13–17                | `"Teen"`        |
#  18–64                | `"Adult"`       |
#  ≥ 65                 | `"Senior"`      |
# Nếu người dùng nhập sai, hãy in "Lỗi cú pháp"
#                                       Giải:
def computeage(age):
    if age < 0 or age > 120:
        return "Invalid age"
    elif age <= 5:
        return "Toddler"
    elif age <= 12:
        return "Child"
    elif age <= 17:
        return "Teen"
    elif age <= 64:
        return "Adult"
    else:
        return "Senior"
try:
    age = int(input("Nhập tuổi: "))
    category = computeage(age)
    print("Category:", category)
except:
    print("Lỗi cú pháp")
```

```python
def max_list(lst):
    return max(lst)  

n = int(input("Nhập số lượng phần tử: "))

lst = []
for i in range(n):
    lst.append(int(input(f"Nhập số thứ {i+1}: ")))

print("Số lớn nhất là:", max_list(lst))
```
