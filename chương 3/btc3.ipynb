```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

if hours > 40:
    pay = (40 * rate) + ((hours - 40) * (rate * 1.5))
else:
    pay = hours * rate

print("Pay:", pay)
```

```python
inp = input("Enter score: ")

try:
    score = float(inp)
except:
    print("Bad score")
    quit()

if score < 0.0 or score > 1.0:
    print("Bad score")
elif score >= 0.9:
    print("A")
elif score >= 0.8:
    print("B")
elif score >= 0.7:
    print("C")
elif score >= 0.6:
    print("D")
else:
    print("F")
```
