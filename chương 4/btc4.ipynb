```python
# Ex 6
def computepay(hours, rate):
    if hours > 40:
        pay = (hours-40)*rate*1.5 + 40*rate
    else:
        pay = hours*rate
    return pay
hours = float(input("Enter hours: "))
rate = float(input("Enter rate: "))
x = computepay(hours, rate)
print("Pay: ", x)
```

```python
# Ex 7
def computegare(score):
    if score < 0.0 or score > 1.0:
        return "Bad score"
    elif score > 0.9:
        return "A"
    elif score > 0.8:
        return "B"
    elif score > 0.7:
        return "C"
    elif score > 0.6:
        return "D"
    else:
        return "F"
try:
    score = float(input("Enter Score: "))
    y = computegare(score)
    print(y)
except:
    print("Bad score")
```
