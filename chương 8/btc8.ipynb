```python
filename = input("Enter file: ")
try:
    file = open(filename)
except:
    print("File not found")
    quit()

my_list=[]

for line in file:
    sentence = line.split()
    for word in sentence:
        if word not in my_list:
            my_list.append(word)
my_list.sort() 
print(my_list)
```

```python
filename = input("Enter a file name: ")

try:
    file = open(filename)
except:
    print("File not found")
    quit()

count = 0

for line in file:

    if not line.startswith('From '): 
        continue
    
    words = line.split()
    
    print(words[1])
    count = count + 1

print("There were", count, "lines in the file with From as the first word")
```

```python
numberlist = []

while True:
    value = input("Enter a number: ")

    if value == "done":
        break

    try:
        
        number = float(value) 
    except:
        print("Invalid input")
        continue

   
    numberlist.append(number)

if len(numberlist) > 0:
    print("Maximum:", max(numberlist))
    print("Minimum:", min(numberlist))
else:
    print("No numbers were entered")
```
