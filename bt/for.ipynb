```python
total=0
count=0
while True:
    value= input("Enter number:")
    if value == 'done':
        break
    try:
        number= int(value)
        total=total + number
        count=count+1
    except:
        print("number not correct")
        continue
if count>0:
    average= total/count
    print("average is",average)
else:
    print("value not")
```
