```python
# Chương 2
# Bài 1:
#  
# Viết chương trình Python yêu cầu người dùng nhập:
# Cân nặng (kg)
# Chiều cao (m)
# Sau đó chương trình:
# Tính chỉ số BMI theo công thức:
# BMI = weight / (height * height)
# In ra BMI đã làm tròn 2 chữ số.
# In ra loại thể trạng dựa trên BMI
#                                     Giải:
weight = float(input("Enter your weight (kg): "))
height = float(input("Enter your height (m): "))

bmi = weight / (height ** 2)
print("Your BMI:", round(bmi, 2))

if bmi < 18.5:
    print("Category: Underweight")
elif bmi < 25:
    print("Category: Normal")
elif bmi < 30:
    print("Category: Overweight")
else:
    print("Category: Obese")
```

```python
# Bài 2:
# Viết chương trình Python yêu cầu người dùng nhập:
# Chiều rộng hình chữ nhật (r)
# Chiều dài hình chữ nhật (h)
# Sau đó chương trình:
# Tính chu vi hình chữ nhật:
# Chu vi = 2 * (r + h)
# Tính diện tích hình chữ nhật:
# Diện tích = r * h
# In ra chu vi và diện tích, làm tròn 2 chữ số thập phân.
#                                      Giải:
r = float(input("Nhập chiều rộng: "))
h = float(input("Nhập chiều dài: "))
chuvi = 2 * (r + h)
dientich = r * h
print("Chu vi hình chữ nhật: ", round(chuvi, 2))
print("Diện tích hình chữ nhật: ", round(dientich, 2))
```

```python
d = float(input("Nhập chiều dài d: "))
r = float(input("Nhập chiều rộng r: "))

dien_tich = d * r
print("Diện tích hình chữ nhật =", dien_tich)
```
