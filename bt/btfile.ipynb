```python
# đọc nội dung file
filename = input("Nhập tên file cần đọc: ")

try:
    file = open(filename)
    content = file.read()
    file.close()

    print("\n--- Nội dung file ---")
    print(content)

except:
    print(" File không tồn tại!")
```

```python
#Đếm số từ trong file
filename = input("Nhập tên file cần đếm từ: ")

try:
    file = open(filename)
    text = file.read()
    file.close()

    words = text.split()
    print("Số từ trong file là:", len(words))

except:
    print("File không tồn tại!")
```

```python
#thay thế từ trong file
file = input("Nhập tên file nguồn: ")
replace_file = input("Nhập tên file kết quả: ")

word = input("Nhập từ cần thay thế: ")
replace_word = input("Nhập từ thay thế: ")

try:
    f = open(file)
    content = f.read()
    f.close()

    new_content = content.replace(word, replace_word)

    f = open(replace_file)
    f.write(new_content)
    f.close()

    print(" Đã thay thế và lưu vào", replace_file)

except:
    print(" File không tồn tại!")
```
